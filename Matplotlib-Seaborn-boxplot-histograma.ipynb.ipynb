```python
# importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#%matplotlib inline
```

```python
# carregando a base de dados
df = pd.read_csv("SINASC_RO_2019.csv")
```

```python
# checando as 5 primeiras linhas da base de dados
df.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Definir tamanho da figura
plt.rc('figure', figsize=(8, 6))

# Definir quantidade de linhas e colunas da subfigura
fig, axes = plt.subplots(1, 2)

# Definir conteúdo de cada subfigura
sns.boxplot(y="IDADEMAE", data=df, ax=axes[0])
df[['IDADEMAE']].boxplot(ax=axes[1])

# Adicionar título aos subplots
axes[0].set_title('Boxplot de Idade da Mãe (Seaborn)')
axes[1].set_title('Boxplot de Idade da Mãe (Matplotlib)')

# Exibir figura
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# Definir tamanho da figura
plt.rc('figure', figsize=(8, 6))

# Definir quantidade de linhas e colunas da subfigura
fig, axes = plt.subplots(1, 2)

# Definir conteúdo de cada subfigura
sns.boxplot(y="PESO", data=df, ax=axes[0])
df[['PESO']].boxplot(ax=axes[1])

# Adicionar título aos subplots
axes[0].set_title('Boxplot do peso do bebê (Seaborn)')
axes[1].set_title('Boxplot do peso do bebê (Matplotlib)')

# Exibir figura
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Definir tamanho da figura
plt.rc('figure', figsize=(8, 6))

# Definir quantidade de linhas e colunas da subfigura
fig, axes = plt.subplots(1, 2)

# Definir conteúdo de cada subfigura
sns.histplot(x="IDADEMAE", data=df, ax=axes[0],
             kde=True, alpha = .25, bins = 20,
             element = 'step'
            )
df['IDADEMAE'].plot.hist(ax=axes[1],bins = 20)

# Adicionar título aos subplots
axes[0].set_title('Histograma de Idade da Mãe (Seaborn)')
axes[1].set_title('Histograma de Idade da Mãe (Matplotlib)')

# Exibir figura
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
# Definir tamanho da figura
plt.rc('figure', figsize=(8, 10))

# Definir quantidade de linhas e colunas da subfigura
fig, axes = plt.subplots(1, 2)

# Definir conteúdo de cada subfigura
sns.histplot(x="PESO", data=df, ax=axes[0])
df[['PESO']].hist(ax=axes[1])

# Adicionar título aos subplots
axes[0].set_title('Histograma do peso do bebê (Seaborn)')
axes[1].set_title('Histograma do peso do bebê (Matplotlib)')

# Adicionar rótulo ao eixo x
axes[0].set_xlabel('Peso (g)')
axes[1].set_xlabel('Peso (g)')


# Exibir figura
plt.show()
```
